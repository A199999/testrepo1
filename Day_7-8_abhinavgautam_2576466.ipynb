```python
#question 1
input_list = []

found = False

match_item = input("Enter the item to be searched: ")

while True:
    element = input("Enter an element (or 'done' to finish): ")
    if element == 'done':
        break
    input_list.append(element)

for i in range(len(input_list)):
    if input_list[i] == match_item:
        found = True
        position = i
        break

if found:
    print(f"The item '{match_item}' was found at position {position}.")
else:
    print(f"The item '{match_item}' was not found in the list.")
```

```python
#question 2
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    position = -1  

    while left <= right:
        mid = (left + right) // 2  

        if arr[mid] == target:
            position = mid  
            break
        elif arr[mid] < target:
            left = mid + 1 
        else:
            right = mid - 1  

    return position

num_inputs = int(input("Enter the number of inputs: "))

input_list = []
for i in range(num_inputs):
    element = input(f"Enter element {i + 1}: ")
    input_list.append(element)

input_list.sort()

target = input("Enter the target value to search: ")

position = binary_search(input_list, target)

if position != -1:
    print(f"The target value '{target}' was found at position {position}.")
else:
    print(f"The target value '{target}' was not found in the list.")
```

```python
#question 3
def selection_sort(arr):
    sorted_list = []  

    while arr:  
        min_index = 0  

        for i in range(1, len(arr)):
            if arr[i] < arr[min_index]:
                min_index = i

        sorted_list.append(arr.pop(min_index))

    return sorted_list

input_list = input("Enter a list of elements separated by spaces: ").split()

input_list = [int(x) for x in input_list]

sorted_result = selection_sort(input_list)

print("Sorted List:", sorted_result)
```

```python
#question 4
def insertion_sort(arr):
    sorted_list = []  

    for i in arr:
        if not sorted_list:  
            sorted_list.append(i)
        else:
            for j in range(len(sorted_list)):
                if i < sorted_list[j]:
                    sorted_list.insert(j, i)
                    break
            else:
                sorted_list.append(i)

    return sorted_list

input_list = input("Enter a list of elements separated by spaces: ").split()

input_list = [int(x) for x in input_list]

sorted_result = insertion_sort(input_list)

print("Sorted List:", sorted_result)

```

```python
#question5
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0  

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

input_list = input("Enter a list of numbers separated by spaces: ").split()

input_list = [int(x) for x in input_list]

merge_sort(input_list)

print("Sorted List:", input_list)
```

```python
#question6
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def find_smallest(self):
        if not self.head:
            return None
        smallest = self.head.data
        current = self.head.next
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest

    def insert_unique(self, data):
        if not self.head:
            self.head = Node(data)
            return
        current = self.head
        while current:
            if current.data == data:
                return  
            if not current.next:
                current.next = Node(data)
                return
            current = current.next

    def display_reverse(self):
        def _display_reverse(node):
            if node:
                _display_reverse(node.next)
                print(node.data, end=" ")

        _display_reverse(self.head)

linked_list = LinkedList()
linked_list.append(5)
linked_list.append(2)
linked_list.append(8)
linked_list.append(2)
linked_list.append(3)

smallest = linked_list.find_smallest()
if smallest is not None:
    print(f"The smallest element in the list is: {smallest}")
else:
    print("The list is empty.")

linked_list.insert_unique(4)

print("Elements in reverse order:")
linked_list.display_reverse()

```

```python
#question7
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            print("Stack is empty. Cannot pop.")

    def display(self):
        if not self.is_empty():
            print("Stack contents:")
            for item in reversed(self.items):
                print(item)
        else:
            print("Stack is empty.")

stack = Stack()

while True:
    print("\nStack Operations:")
    print("1. Push")
    print("2. Pop")
    print("3. Display")
    print("4. Quit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == '1':
        item = input("Enter item to push: ")
        stack.push(item)
        print(f"{item} pushed onto the stack.")

    elif choice == '2':
        popped_item = stack.pop()
        if popped_item is not None:
            print(f"Popped item: {popped_item}")

    elif choice == '3':
        stack.display()

    elif choice == '4':
        print("Exiting the program.")
        break

    else:
        print("Invalid choice. Please choose from 1/2/3/4.")
```

```python
#question8
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def insert(self, item):
        self.items.append(item)

    def delete(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            print("Queue is empty. Cannot delete.")

    def display(self):
        if not self.is_empty():
            print("Queue contents:")
            for item in self.items:
                print(item)
        else:
            print("Queue is empty.")

queue = Queue()

while True:
    print("\nQueue Operations:")
    print("1. Insert")
    print("2. Delete")
    print("3. Display")
    print("4. Quit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == '1':
        item = input("Enter item to insert: ")
        queue.insert(item)
        print(f"{item} inserted into the queue.")

    elif choice == '2':
        deleted_item = queue.delete()
        if deleted_item is not None:
            print(f"Deleted item: {deleted_item}")

    elif choice == '3':
        queue.display()

    elif choice == '4':
        print("Exiting the program.")
        break

    else:
        print("Invalid choice. Please choose from 1/2/3/4.")
```

```python
#question9
def infix_to_postfix(infix_expr):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}

    def is_operator(token):
        return token in "+-*/^"

    def higher_precedence(op1, op2):
        return precedence[op1] > precedence[op2]

    stack = []
    postfix_expr = []
    
    for token in infix_expr:
        if token.isalnum():  # Operand
            postfix_expr.append(token)
        elif token == '(':  # Left parenthesis
            stack.append(token)
        elif token == ')':  # Right parenthesis
            while stack and stack[-1] != '(':
                postfix_expr.append(stack.pop())
            stack.pop()  # Remove the '(' from the stack
        elif is_operator(token):
            while stack and stack[-1] != '(' and higher_precedence(stack[-1], token):
                postfix_expr.append(stack.pop())
            stack.append(token)

    while stack:
        postfix_expr.append(stack.pop())

    return ''.join(postfix_expr)
infix1 = "ABC-DE-FG"
infix2 = "(B^2-4*AC/X/2)(100)"

postfix1 = infix_to_postfix(infix1)
postfix2 = infix_to_postfix(infix2)

print("Infix 1:", infix1)
print("Postfix 1:", postfix1)

print("\nInfix 2:", infix2)
print("Postfix 2:", postfix2)
```
